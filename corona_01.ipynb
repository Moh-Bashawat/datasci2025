```python
# drop rows with too many missing values
df = df.dropna(thresh=10) # keep rows with at least 10 non-NA values
```

```python
# fill missing numeric values with median:
df_new= df.copy()
df_new['total_cases'] = df_new['total_cases'].fillna(df_new['total_cases'].median())
```

```python
# fix data types
df['date'] = pd.to_datetime(df.loc[:, 'date'])
```

```python
#identify and handle outliers (example for new_cases):
import numpy as np
z_score = np.abs(df['new_cases'] - df['new_cases'].mean()) / (df['new_cases'].std())
```

```python
df_new = df[z_score < 3]
```
